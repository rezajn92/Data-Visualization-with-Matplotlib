## Exercise 3

** Creating the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,.4,.4])
```

** Using x,y, and z arrays to recreate the plot below. **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel ('X')
ax2.set_ylabel ('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

## Exercise 4

** Using plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig=plt.subplots(nrows=1, ncols=2)
```

** Plotting (x,y) and (x,z) on the axes.**

```python
fig,ax=plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y, linestyle='--', linewidth='1')
ax[1].plot(x,z, marker='+', linewidth='1')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].plot(x,y, linestyle='--', linewidth='1')
ax[1].plot(x,z, marker='+', linewidth='1')
plt.tight_layout()
```
